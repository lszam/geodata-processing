## 3. Definir rea de interesse
H muitas formas de definir a rea de interess para o estudo. Aqui, ser atravs de um ponto central, e um buffer entorno deste ponto.

> Minha rea de interesse  Cachoeria Paulista , um municpio da regio do Vale do Paraba, no interior do estado de So Paulo. Localiza-se a uma latitude 2239'54" sul e a uma longitude 4500'34" oeste, estando a uma altitude de 521 metros."






```python
# Regio de interesse
ponto = ee.Geometry.Point ([-45.00, -22.66])  # Primeiro longitude, depois latitude
# O valor passado para a funo buffer  o raio em metros
raio = 10000  # km
area = ponto.buffer(raio).bounds()

# Plotar mapa
Map = geemap.Map(center=ponto.coordinates().getInfo()[::-1], zoom=10)  # Inverta a ordem das coordenadas para o parmetro center
Map.addLayer(area, {}, 'Regio de Interesse')
Map
```

Catlogo do EE

> *Meu dado escolhido: USGS Landsat 8 Collection 2 Tier 1 Raw Scenes.*

Solicito aqui para pegar a primeira imagem que atenda aos critrios, pelo mtodo .first()

```python
# Solicito aqui para pegar a primeira imagem que atenda aos critrios
# pelo mtodo .first()

collection = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2')\
.filterBounds(area) \
    .filter(ee.Filter.lt('CLOUD_COVER', 1)) \
    .filterDate('2024-01-01','2024-07-01') \
    .sort('CLOUD_COVER') \
    .first() #pega a primeira imagem que atenda aos critrios
```

Vamos usar a composio de bandas **['SR_B4', 'SR_B3', 'SR_B2']**, para cores naturais.

Plotando a composio sem nenhum tratamento dos dados, a imagem geralmente  de difcil visualizao

```python
#Plotar - MAPA INTERATIVO
Map = geemap.Map(center=ponto.coordinates().getInfo()[::-1], zoom=12)  # Inverte a ordem das coordenadas para o parmetro center
Map.addLayer(collection, {'bands': ['SR_B4', 'SR_B3', 'SR_B2']}, 'LANDSAT/LC08/C02/T1_L2')
Map.addLayer(area, {}, 'Regio de Interesse')
Map
```

```python
import matplotlib.pyplot as plt

#Plotando no modo do Matplotlib
banda_R, banda_G, banda_B = collection.select('SR_B4'), collection.select('SR_B3'), collection.select('SR_B2')

# Obter a imagem como um array NumPy, jah usando a variaveil area como limite
R_array = geemap.ee_to_numpy(banda_R, region=area)
G_array = geemap.ee_to_numpy(banda_G, region=area)
B_array = geemap.ee_to_numpy(banda_B, region=area)

# Exibir a imagem usando Matplotlib com histogramas
fig, axs = plt.subplots(2, 3, figsize=(12, 6))

# Imagens das bandas
axs[0, 0].imshow(R_array, cmap='gray')
axs[0, 0].set_title('Banda R')
axs[0, 0].axis('off')

axs[0, 1].imshow(G_array, cmap='gray')
axs[0, 1].set_title('Banda G')
axs[0, 1].axis('off')

axs[0, 2].imshow(B_array, cmap='gray')
axs[0, 2].set_title('Banda B')
axs[0, 2].axis('off')

# Histogramas das bandas
axs[1, 0].hist(R_array.ravel(), bins=256, color='red', alpha=0.75)
axs[1, 0].set_title('Histograma R')

axs[1, 1].hist(G_array.ravel(), bins=256, color='green', alpha=0.75)
axs[1, 1].set_title('Histograma G')

axs[1, 2].hist(B_array.ravel(), bins=256, color='blue', alpha=0.75)
axs[1, 2].set_title('Histograma B')

plt.tight_layout()
plt.show()
```

## 5. Processar valores para otimizar a visualizao
####Intervalo de valores
Para distribuir melhro a intensidade das cores, vamos calcular os valores de percentil do conjunto de bandas escolhido. O objetivo  escolher um intervalo de valores mais adequado para a visualizao.

>Percentil: o dado  separado em 100 partes, do menor para o maior valor. O 2 percentil equivale aos 2% menores valores do conjunto de dados; e o 98 percentil determina o limite dos 98% menores valores. Mais sobre em https://pt.wikipedia.org/wiki/Percentil#:~:text=Em%20estat%C3%ADstica%20descritiva%2C%20os%20percentis,N%20%C3%A9%20o%20tamanho%20amostral.

Aqui vamos usar dados entre os percentis 2 e 98 para definir o intervalo de valores.

```python
import numpy as np

# Calcular os percentis para as bandas
percentiles = collection.reduceRegion(
    reducer=ee.Reducer.percentile([2, 98]),
    geometry=area,
    scale=30,
    maxPixels=1e9
).getInfo()

# Extrair os valores de percentil
min_values = [percentiles[f"SR_B{band}_p2"] for band in [4, 3, 2]]
max_values = [percentiles[f"SR_B{band}_p98"] for band in [4, 3, 2]]
print('Min values:', min_values)
print('Max values:', max_values)

# Definir os parmetros de visualizao com valores ajustados manualmente
viz_params = {
    'bands': ['SR_B4', 'SR_B3', 'SR_B2'],  # Bandas para Red, Green, Blue
    'min': np.min(min_values),
    'max': np.max(max_values),  # Ajustar manualmente para uma melhor visualizao
    'gamma': 1.2  # para melhorar a visualizao
}
```

###Correo de histograma
A correo utilizando o histograma da imagem  usada para melhorar a visualizao.
  - Equalizao de Histograma: transformao matemtica que resulta da distribuio uniforme de valores dos pixels. Auxilia na visualizao de detalhes que podem estar sub-representados nas imagens originais.
  - Contraste: com a redistribuio de valores, a equalizao do histograma melhora o contraste da imagem. Isto significa que as diferenas nas intensidades ficaro mais perceptveis, o que ajuda na deteco de bordas.

> Adicionalmente, este tipo de correo contribui para o uso em anlises automatizadas. O ajuste do contraste e das distribuies de valores contribui par ao sucesso de tcnicas de classificao automatizada.

#### Aplicar correo de histograma

```python
# CALCULAR HISTOGRAMA E CORRECAO

# Calcular os valores percentis
percentiles = collection.reduceRegion(
    reducer=ee.Reducer.percentile([2, 98]),
    geometry=area,
    scale=30,
    maxPixels=1e9
).getInfo()

# Extrair os valores
min_values = [percentiles[f"SR_B{band}_p2"] for band in [4, 3, 2]]
max_values = [percentiles[f"SR_B{band}_p98"] for band in [4, 3, 2]]

# Funo para aplicar correo de histograma
def histogram_equalization(image, bands):
    equalized = image.select(bands).unitScale(min(min_values), max(max_values)).multiply(255).toByte()
    return equalized

# Aplicar correo de histograma  imagem
equalized_image = histogram_equalization(collection, ['SR_B4', 'SR_B3', 'SR_B2'])
```

```python
# VISUALIZAR HISTOGRAMAS APOS A CORRECAO

#Plotando no modo do Matplotlib
banda_R, banda_G, banda_B = equalized_image.select('SR_B4'), equalized_image.select('SR_B3'), equalized_image.select('SR_B2')

# Obter a imagem como um array NumPy, jah usando a variaveil area como limite
R_array = geemap.ee_to_numpy(banda_R, region=area)
G_array = geemap.ee_to_numpy(banda_G, region=area)
B_array = geemap.ee_to_numpy(banda_B, region=area)

# Exibir a imagem usando Matplotlib com histogramas
fig, axs = plt.subplots(2, 3, figsize=(12, 6))

# Imagens das bandas
axs[0, 0].imshow(R_array, cmap='gray')
axs[0, 0].set_title('Banda R')
axs[0, 0].axis('off')

axs[0, 1].imshow(G_array, cmap='gray')
axs[0, 1].set_title('Banda G')
axs[0, 1].axis('off')

axs[0, 2].imshow(B_array, cmap='gray')
axs[0, 2].set_title('Banda B')
axs[0, 2].axis('off')

# Histogramas das bandas
axs[1, 0].hist(R_array.ravel(), bins=256, color='red', alpha=0.75)
axs[1, 0].set_title('Histograma R')

axs[1, 1].hist(G_array.ravel(), bins=256, color='green', alpha=0.75)
axs[1, 1].set_title('Histograma G')

axs[1, 2].hist(B_array.ravel(), bins=256, color='blue', alpha=0.75)
axs[1, 2].set_title('Histograma B')

plt.tight_layout()
plt.show()
```
