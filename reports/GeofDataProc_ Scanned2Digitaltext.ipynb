```python
#Configurar a variável ambiente da bib que trata o texto
import os
os.environ["TESSDATA_PREFIX"] = "/usr/share/tesseract-ocr/4.00/tessdata/"

import pytesseract
from pdf2image import convert_from_path
import docx

#------------------------------------------------------------------------------#

# Input
pdf_path = "/content/Relatório Final - São Paulo - Rio de Janeiro - 1039.pdf"

# Escolher o intervalo de páginas para converter
images = convert_from_path(pdf_path, first_page=1, last_page=20)

# Função para limpar caracteres não suportados pela biblioteca
def clean_text(text):
    return ''.join(c for c in text if c.isprintable())

# Ler imagens e concatenar o texto
text = ""
for i, image in enumerate(images):
    page_text = pytesseract.image_to_string(image, lang='por')  # português
    page_text = clean_text(page_text)  # Limpar o texto
    text += f"--- Página {i + 1} ---\n"
    text += page_text + "\n\n"

# Criar documento Word
doc = docx.Document()
doc.add_paragraph(text)
output_path = "/content/Relatório_Final_São_Paulo_Rio_de_Janeiro_1039_PYTHON.docx"
doc.save(output_path)

print(f"Documento salvo em: {output_path}")
```
